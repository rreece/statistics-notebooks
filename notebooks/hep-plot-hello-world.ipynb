```python
import math
import hepplot as hep
```

```python
bins = list(range(6))
y = [ [1,4,6,8,9], [2,2,7,10,12], [2,8,5,10,6] ]
data = [6, 12, 19, 30, 26]
labels = ['A','B','C']

data.reverse()
for y_i in y:
    y_i.reverse()

ytotal  = [sum(i) for i in zip(*y)]
yerr    = [
    [0.2*hep.stat.poisson_error_down(_y) for _y in ytotal],
    [0.2*hep.stat.poisson_error_up(_y) for _y in ytotal],
]
yerrs   = [
    [0.3*math.sqrt(_y) for _y in y_i] for y_i in y
]
```

Plot a 1-d histogram with a multi-component model and data.

```python
hep.plot.hist1d(bins, y, yerr=None, data=data, labels=labels)
```

Plot a 1-d histogram with a multi-component model, data, and the data/model ratio.

```python
hep.plot.hist1d(bins, y, yerr=None, data=data, labels=labels, ratio=True)
```

Plot a 1-d histogram with a multi-component model, model uncertainty, data, and the data/model ratio.

```python
hep.plot.hist1d(bins, y, yerr=yerr, data=data, labels=labels, ratio=True)
```

Plot a 1-d histogram with a multi-component model, multi-component model uncertainty, data, and the data/model ratio.

```python
hep.plot.hist1d(bins, y, yerrs=yerrs, data=data, labels=labels, ratio=True)
```
